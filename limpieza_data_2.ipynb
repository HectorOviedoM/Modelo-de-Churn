Los outliers de las columnas TRD_IM van a ser aplanados utilizando el criterio del rango intercuartilico

```python
def iqr_superior(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_high = q3+(1.5*iqr)
    return fence_high
```

```python
sup_1 = iqr_superior(data_2, 'TRD_IM_M1')
sup_2 = iqr_superior(data_2, 'TRD_IM_M2')
sup_3 = iqr_superior(data_2, 'TRD_IM_M3')
```

```python
#Aplanamiento usando el criterio del iqr superior 

data_2.loc[data_2['TRD_IM_M1']>sup_1,'TRD_IM_M1'] = sup_1

data_2.loc[data_2['TRD_IM_M2']>sup_2,'TRD_IM_M2'] = sup_2

data_2.loc[data_2['TRD_IM_M3']>sup_3,'TRD_IM_M3'] = sup_3
```
